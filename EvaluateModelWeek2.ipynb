# Regression Predictive Model

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/slr.csv')

# View the first few rows
data.head()
```

 Define  historical input feature X : the chirps per second and the output feature y: temperature (F) in pandas

```python
X = data[['chirps per second']]
y = data['temperature (F)']
#Note: The double brackets around ‘chirps per second’ are there because we will later be ingesting our X and y into scikit-learn. Scikit-learn can take any number of predictor features X and it expects a 2-D array for X which is why we need the double brackets.
```

 partition our historical data into a train set and a test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42,
test_size = 0.25)
```

Check the size of your whole data and the size of training and test sets using the shape function.

```python
print('Whole Data shape', data.shape)
print('X_train shape', X_train.shape)
print('X_test shape', X_test.shape)
```

Single Linear Regression (SLR) model is to instantiate it.
This creates a blank model object (algorithm) which is much like an empty template ready to be filled (fitted)
with our training data.

```python
from sklearn.linear_model import LinearRegression
slr_model = LinearRegression()
```

## Fit the Model on Train Set:

```python
slr_model.fit(X_train, y_train)
```

The slope and the y-intercept for  model and then view a scatterplot with the actual regression line

```python
#To see the simple linear regression parameters
slope = slr_model.coef_
y_intercept = slr_model.intercept_
print('Slope', slope)
print('Intercept', y_intercept)
```

## Make Predictions on Train Set

```python
y_pred_train = slr_model.predict(X_train)
```

Compare our actual y from the train set (y_train) to the predicted values
of y ( y_pred_train) by making use of the R2 metric and the MSE metrics

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_train = r2_score(y_train, y_pred_train)
mse_train = mean_squared_error(y_train, y_pred_train)
print('Training R2 score', r2_train)
print('Training MSE', mse_train)
```

The training **R2** value is 0.70. This means that the model we just built explains about 70% of the variability in our
data. The training **MSE** is 13.77 which means our predictions on average deviate from the true values of y by about
3.71 degrees of Fahrenheit. The MSE is the mean *squared* error, so to get the actual error with the same units as
our **y** (deg. Fahrenheit) we need to take the ***square root of the MSE***.

## Make predictions on the test data

```python
y_pred_test = slr_model.predict(X_test)
```

### Evaluate the model’s performance on Test Set

```python
r2_test = r2_score(y_test, y_pred_test)
mse_test = mean_squared_error(y_test, y_pred_test)
print('Test R2 score', r2_test)
print('Test MSE', mse_test)
```

Compare the test instances y_test and y_pred_test in comparison_df.

```python
Comparison_df = pd.DataFrame({'Actual' : y_test, 'Predicted' : y_pred_test})
Comparison_df.to_csv(r'/content/Comparison_df.csv', index=True)
Comparison_df
```

the type of Wine.</li>

product rating from 1 to 5.</li></ol>
